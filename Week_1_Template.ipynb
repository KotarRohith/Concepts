### b. In 2020, what month had the hottest average temperature? The coldest?

```python
wdf['year'] = wdf['date'].dt.year
wdf['month'] = wdf['date'].dt.month
wdf['dt'] = pd.to_datetime(wdf['date']).dt.date
y2020=wdf[wdf['year']==2020]
avgtemp=y2020.groupby(y2020['month']).agg({"tmp_meas": "mean"})
print("Hottest Month",avgtemp[avgtemp['tmp_meas']==avgtemp['tmp_meas'].max()])
print("Coldest Month",avgtemp[avgtemp['tmp_meas']==avgtemp['tmp_meas'].min()])

#print("Hottest Month",wdf[wdf['tmp_meas']==wdf['tmp_meas'].max()]['month'])
```
